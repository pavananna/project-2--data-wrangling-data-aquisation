```python
import seaborn as sns #to find yhe outliers
import numpy as np
```

```python
new_dataset.dtypes
```

```python
numeric_columns=new_dataset[['temp','atemp','hum','windspeed']]  #these are the columns that we can find the outliers.
sns.boxplot(numeric_columns)   #box_plot
```

```python
for i in numeric_columns:
  Q1=np.quantile(new_dataset[i],0.25)
  Q3=np.quantile(new_dataset[i],0.75)

  IQR= Q3-Q1

  upper_limit = Q3 + (1.5 * IQR)

  lower_limit= Q1 - (1.5 * IQR)

  new_dataset.loc[new_dataset[i] > upper_limit , i ]=upper_limit
  new_dataset.loc[new_dataset[i] < lower_limit ,i ]=lower_limit
```
